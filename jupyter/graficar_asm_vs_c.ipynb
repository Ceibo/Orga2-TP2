```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
from sklearn import linear_model
import random

import warnings
warnings.filterwarnings('ignore')
```

```python
# Levanto los datos

df = pd.read_csv('solver_project_c_vs_asm_3.csv')
```

```python
print("Decripción de los datos")
print(df.describe())

print("\nLas primeras 6 filas de los datos")
print(df.head(6))
```

```python
# Gráficos los datos
sns.set(font_scale=1.9) 
sns.tsplot(time=df['tamanio'], data=df['data_c'],condition=u"Tiempos C")#, interpolate=False)
sns.tsplot(time=df['tamanio'], data=df['data_asm'], color='r',condition=u"Tiempos ASM")#, interpolate=False)
sns.plt.ylabel(u'Ticks', fontsize=24, fontweight='bold')
sns.plt.xlabel(u'Tamaño', fontsize=24, fontweight='bold')
sns.plt.yscale('log')
sns.plt.xscale('log')
sns.plt.xlim(0,1000 )
sns.plt.show()
```
